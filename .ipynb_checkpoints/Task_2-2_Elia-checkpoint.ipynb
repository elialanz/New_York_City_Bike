### 4. In a new notebook, import all necessary libraries, read in your data, and join it. Hint: what’s the most effective way to import and join data in such a format?

```python
# Import Libraries

import pandas as pd 
import numpy as np
import os 
import requests
import json
from datetime import datetime
```

```python
# Define the folder path where the CityBike data files are stored

folderpath = r"CB_Data"
```

```python
# Create full file paths for all files inside the folder

filepaths  = [os.path.join(folderpath, name) for name in os.listdir(folderpath)]
```

```python
# Display the list of full file paths found in the folder

filepaths
```

```python
# Load and combine all CSV files into one dataframe (as strings)

df = pd.concat(
    [pd.read_csv(f, dtype=str) for f in filepaths],
    ignore_index=True
)
```

```python
# Read each CSV separately into a list of dataframes

dfs = [pd.read_csv(f, dtype=str) for f in filepaths]
```

```python
# Print number of rows in each CSV to verify consistency across files

print("Row counts per file:")
for path, d in zip(filepaths, dfs):
    print(path, "->", d.shape)
```

```python
# Concatenate all individual dataframes into one combined dataframe

df = pd.concat(dfs, ignore_index=True)
```

```python
# Verify final row count matches the sum of all input CSVs

print("\nFinal DF shape:", df.shape)
print("Total rows expected:", sum(len(d) for d in dfs))
print("Match:", len(df) == sum(len(d) for d in dfs))
```

```python
# Check that all CSV files contain the same column names

for path, d in zip(filepaths, dfs):
    print("\n", path)
    print(sorted(d.columns))
```

```python
# Inspect dataframe structure and preview first/last rows

df.info()
df.head()
df.tail()
```

#### **I am going to create a column that calculates trip duration, as it is not present and it may be useful to have it later on.**

```python
# Convert start and end timestamps to datetime, coercing invalid values

df['started_at'] = pd.to_datetime(df['started_at'], errors='coerce')
df['ended_at']   = pd.to_datetime(df['ended_at'], errors='coerce')
```

```python
# Calculate trip duration in minutes from start and end timestamps

df['tripduration'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60
```

```python
# Get the API

r = requests.get('https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GHCND&datatypeid=TAVG&limit=1000&stationid=GHCND:USW00014732&startdate=2022-01-01&enddate=2022-12-31', headers={'token':Token})
```

```python
# Load the api response as a json

d = json.loads(r.text)
```

```python
# Just a quick look

d
```

```python
# Extract only the average-temperature records (datatype = 'TAVG')

avg_temps = [item for item in d['results'] if item['datatype']=='TAVG']
```

```python
# Extract the date value from each average-temperature record

dates_temp = [item['date'] for item in avg_temps]
```

```python
# Extract the temperature value from each average temperature record

temps = [item['value'] for item in avg_temps]
```

```python
# Just a quick look

temps
```

```python
# Put the results in a dataframe

df_temp = pd.DataFrame()
```

```python
# Convert temperature record dates from string format to datetime objects

df_temp['date'] = [datetime.strptime(d, "%Y-%m-%dT%H:%M:%S") for d in dates_temp]
```

```python
# Convert temperature values to Celsius by dividing by 10

df_temp['avgTemp'] = [float(v)/10.0 for v in temps]
```

### 8. Merge the weather data with the New York CitiBike data set and export it to a CSV file.

```python
# Check dataframe data types

df.dtypes
```

```python
# Ensure "started_at" and "ended_at" columns are parsed as datetime objects

df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
```

```python
# Extract just the calendar date from the "started_at" timestamp

df['date'] = pd.to_datetime(df['started_at'], format='%Y-%m-%d').dt.date
```

```python
# Convert the date column back to a datetime object for consistency

df['date'] = pd.to_datetime(df['date'])
```
