### Extra Consistency and Cleaning Checks For Datasets

```python
# Import Required Libraries

import pandas as pd
import numpy as np
import matplotlib
import os
from matplotlib import pyplot as plt
from datetime import datetime as dt
```

```python
# Assigning Path

path = r'/Users/elia/Desktop/New_York_City_Bike'
```

```python
# Import Merged CitiBike DataSet as a DataFrame

df = pd.read_csv(os.path.join(path, 'CB_Data', 'Prepared Data', 'ny_citybike.csv'))
```

```python
# Identify columns in df dataframe that contain mixed data types

for col in df.columns.tolist():
  weird = (df[[col]].map(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# Count missing values in each column of the dataframe

df.isnull().sum()
```

```python
# Inspect data types inside the 'start_station_name' column
col = 'start_station_name'

# Count how many values are of each Python type
type_counts = df[col].map(lambda x: type(x).__name__).value_counts(dropna=False)

# Total number of records in the column
total_count = len(df[col])

# Print column name, data type counts, and total number of records
print(f"Column: {col}")
print("\nCounts by Python type:")
print(type_counts)
print("\nTotal records:", total_count)
```

```python
# Specify which column to analyze for data type issues
col = "start_station_name"

# Identify all unique Python types inside the column
unique_types = df[col].map(type).unique()

# Loop through each detected data type, show sample rows, and count occurrences
for t in unique_types:
    print(f"\n=== Data type: {t.__name__} ===")
    subset = df[df[col].map(type) == t]
    display(subset.head(3))
    print(f"Total rows of this type: {len(subset)}")
```
